```python
%%writefile tools.py
def in2ilist():
  return list(map(int,input().split()))

def in2ihw(_h):
  _A=[]
  for _ in range(_h):
    _a=list(map(int,input().split()))
    _A.append(_a)
  return _A

def in2shw(_h):
  _A=[]
  for _ in range(_h):
    _a=input()
    _A.append(_a)
  return _A
```
