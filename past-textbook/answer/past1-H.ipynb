```python
%%writefile atcoder.py
def in2ilist():
  return list(map(int,input().split()))

N = int(input())
C = in2ilist()
Q = int(input())
ans=0
minzC=min(C)
minsC=10000000000
for i in range(N):
  if i%2==0:
    minsC=min(minsC,C[i])

zen=0
set=0
for i in range(Q):
  S=in2ilist()
  if S[0]==1:
    if S[1]%2==0: 
      if C[S[1]-1]-zen>=S[2]:
        ans+=S[2]
        C[S[1]-1]-=S[2]
        minzC=min(minzC,C[S[1]-1])
    else: 
      if C[S[1]-1]-zen-set>=S[2]:
        ans+=S[2]
        C[S[1]-1]-=S[2]
        minsC=min(minsC,C[S[1]-1])
         
    
  
  if S[0]==2:
    if minsC-set-zen>=S[1]:
      set+=S[1]
    
    
  if S[0]==3:
    if min(minzC-zen,minsC-set-zen)>=S[1]:
      zen+=S[1]

if N%2==0:
  ans+=set*int(N/2)
else:
  ans+=set*int((N+1)/2)   
ans+=zen*N
print(ans)

```
