よく使いそうな関数を、Toolsにまとめておきます。コンテスト前にファイルを作っておいて、入れておくと便利です。

```python
%%writefile atcoder.py
from collections import deque
import sys
sys.setrecursionlimit(1000000)

def in2ilist():
  return list(map(int,input().split()))

N,W=in2ilist()
ws=[0]
vs=[0]
for i in range(N):
  _w,_v=in2ilist()
  ws.append(_w)
  vs.append(_v)

value=[]
for i in range(N+1):
  value.append([-10**18]*(W+1))

value[0][0]=0
for i in range(1,N+1):
  for w in range(W+1):
    value[i][w]=max(value[i][w],value[i-1][w])
    if w-ws[i]>=0:
      value[i][w]=max(value[i][w],value[i-1][w-ws[i]]+vs[i])

ans=max(value[N])
print(ans)
```
