```python
%%writefile atcoder.py
from collections import deque
import sys
sys.setrecursionlimit(1000000)

def make_1d_none(_h):
    _a=[]
    for _ in range(_h):
        _a.append([])
    for i in range(1,_h):
      b=int(input())
      _a[b-1].append(i)
    return _a

N = int(input())
child=make_1d_none(N)
####
def dfs(i):
  if len(child[i])==0:
    return 1
  else:
    values=[]
    for j in child[i]:
      values.append(dfs(j))
    return max(values)+min(values)+1


print(dfs(0))
####
```
