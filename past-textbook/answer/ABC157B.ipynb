本文と違い、2次元配列を読み込む専用の関数を準備します。<p>
def in2ihw(_h): <p>

```python
%%writefile atcoder.py
def in2ihw(_h):
  _A=[]
  for _ in range(_h):
    _a=list(map(int,input().split()))
    _A.append(_a)
  return _A

A=in2ihw(3)
n=int(input())
b=[]
m=[[False,False,False],[False,False,False],[False,False,False]]

for _ in range(n):
  _b=int(input())
  b.append(_b)
  for i in range(0,3):
    for j in range(0,3):
      if A[i][j]==_b:
        m[i][j]=True

bingo=False
for i in range(0,3):
  if m[i][0] and m[i][1] and m[i][2]:
    bingo=True
  if m[0][i] and m[1][i] and m[2][i]:
    bingo=True

if m[0][0] and m[1][1] and m[2][2]:
    bingo=True
if m[0][2] and m[1][1] and m[2][0]:
    bingo=True

if bingo:
  print("Yes")
else:
  print("No")
```
