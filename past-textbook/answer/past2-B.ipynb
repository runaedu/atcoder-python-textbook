```python
%%writefile atcoder.py

S=input()
n=len(S)

a=0
b=0
c=0

for i in range(n):
  if S[i]=='a':a+=1
  if S[i]=='b':b+=1
  if S[i]=='c':c+=1

if a>b and a>c: print('a')
if b>a and b>c: print('b')
if c>b and c>a: print('c')
```
