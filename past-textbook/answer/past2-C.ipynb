本文と違い、2次元配列を読み込む専用の関数を準備します。<p>
def in2shw(_h): <p>

```python
%%writefile atcoder.py
def in2shw(_h):
  _A=[]
  for _ in range(_h):
    _a=input()
    _a=list(_a)
    _A.append(_a)
  return _A

n=int(input())
S=in2shw(n)

for i in range(n-2,-1,-1):
  for j in range(1,2*n-2):
    if S[i][j]=="#":
      if S[i+1][j-1]=="X" or S[i+1][j]=="X" or S[i+1][j+1]=='X':
          S[i][j]="X"

for i in range(n):
  S[i]=''.join(S[i])
  print(S[i])
```
