```python
%%writefile atcoder.py
from collections import deque
import sys
sys.setrecursionlimit(1000000)

def in2ilist():
  return list(map(int,input().split()))

N=int(input())
ps=[0]+in2ilist()
P=sum(ps)

exist=[]
for i in range(N+1):
  exist.append([False]*(P+1))

exist[0][0]=True

for i in range(1,N+1):
  for j in range(P+1):
    if exist[i-1][j]:
      exist[i][j]=True
    if j>=ps[i] and exist[i-1][j-ps[i]]:
      exist[i][j]=True

ans=0
for j in range(P+1):
  if exist[N][j]:
    ans+=1

print(ans)
```
